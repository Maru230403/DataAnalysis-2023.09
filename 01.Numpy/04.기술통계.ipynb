### 1.4 기술 통계

```python
import numpy as np
```

```python
x = np.array([18,   5,  10,  23,  19,  -8,  10,   0,   0,   5,   2,  15,   8,
              2,   5,   4,  15,  -1,   4,  -7, -24,   7,   9,  -6,  23, -13])
```

- 데이터의 갯수

```python
x.size, len(x)
```

- 표본 평균

```python
np.mean(x), x.mean()
```

```python
np.sum(x) / x.size
```

- 표본(Sample)분산(Variance)

```python
np.var(x), x.var()
```

```python
# 비편향 분산, 표본 분산은 N - 1로 나누어 준다.
np.var(x, ddof=1)
```

- 표본 표준편차(Standard Deviation)

```python
np.std(x), x.std(), np.sqrt(np.var(x))
```

```python
# 표본집단의 표준편차(비편향 표준편차)
np.std(x, ddof=1), x.std(ddof=1)
```

- 최댓값, 최솟값

```python
np.max(x), x.max(), np.min(x), x.min()
```

- 중앙값(Median)

```python
np.sort(x)
```

```python
np.median(x)
```

- 사분위수(Quartile)

```python
# Box plot
import matplotlib.pyplot as plt
plt.boxplot(x)
plt.show()
```

```python
# 25% ~ 75% :IQR
# 1.5 * IQR의 값만큼 1분위수와 3분위수부터 거리를 늘려 작성(?)
# 데이터를 분포로 그리는것이 아니라 사분위수로 그려서 쉽게 작성할 수 있다.
```

```python
np.percentile(x, 0), np.percentile(x, 25), np.percentile(x, 50), np.percentile(x, 75), np.percentile(x, 100)
```

```python
# 최솟값인 -24.0 은 그림에서 동그라미로 표시되고, 동그라미로 표시된 데이터는
# 이상티(Qutlier)라 함
```
