### 2.6 인덱스 조작

```python
import numpy as np
import pandas as pd
```

- 인덱스 설정 및 제거

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
df2 = df1.set_index('C1')
df2
```

```python
print(df1.index)
print(df2.index)
```

```python
df2.reset_index() # inplace=True가 되는 순간 값이 바뀐다.
```

```python
df1.to_csv('data/sample11.csv', index=False)
pd.read_csv('data/sample11.csv')
```

```python
df2.to_csv('data/sample12.csv')
pd.read_csv('data/sample12.csv',index_col='C1')
```

- 다중 인덱스

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])
df3
```

```python
# 0.41
df3['A','C2'][2]
```

```python
df3['A']
```

```python
df3.columns.names = ['Cidx1','Cidx2']
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
# 0.41
df4['A','D']['M','id_3']
```

```python
# iloc 인덱서를 사용하면 다중 인덱스 무시하고 인덱싱하면 됨
df4.iloc[2, 1]
```
