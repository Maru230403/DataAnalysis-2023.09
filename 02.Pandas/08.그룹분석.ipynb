#### 1. 피벗(pivot) 테이블

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot(values='인구', index='도시', columns='연도')
```

```python
df1.pivot(values='인구', index=['지역','도시'], columns='연도') # 피벗을 사용할때는 앞에 벨류,컬럼,인덱스를 붙여서 사용
```

```python
df1.pivot_table('인구',['지역','도시'],'연도')     #첫번쨰는 벨류 두번쨰는 인덱스 세번쨰는 칼럼 보통 피벗테이블을 많이 사용함.
```

### 2. 그룹 분석

```python
import seaborn as sns
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

- group by

```python
iris.describe()
```

```python
iris.species.value_counts()
```

```python
# select 집계함수 group by  group by가 나올려면  무조건 집계 함수가 나와야한다. (count, mean 등등)
# Iris 품종별 항목의 갯수
iris.groupby('species').count()
```

```python
# Iris 품종별 feature의 평균
iris.groupby('species').mean()
```

```python
# 품종별 sepal(꽃받침)_length(꽃받침 길이)의 평균               petal(꽃잎)
iris.groupby('species').mean()['sepal_length']      # 결과는  Series
```

```python
iris.groupby('species')['sepal_length'].mean()
```

```python
# 품종별 sepal(꽃받침)_length(꽃받침 길이)의 평균을 데이터프레임으로
iris[['sepal_length', 'species']].groupby('species').mean()
```

```python
# 품종별 피처의 평균과 표준편차
iris.groupby('species').agg(['mean','std'])     # 두가지 이상 집계함수를 사용할때는 agg()에 리스트형식으로 함수 이름만 넣으면 된다.
```

```python
# sepal length  품종별 평균, 표준편차, 최댓값, 최솟값
iris.groupby('species')['sepal_length'].agg(['mean','std','max','min'])
```

```python
# tip_pct 컬럼 생성
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
# 성별 팁비율의 평균, 최대, 최저
tips.groupby('sex')['tip_pct'].agg(['mean','max','min'])
```

```python
# 흡연 유무별로 팁비율의 평균, 표준편차
tips.groupby('smoker')['tip_pct'].agg(['mean','std','max','min'])
```

```python
# 성별, 흡연유무별 팁 비율의 평균, 표준편차
tips.groupby(['sex','smoker'])['tip_pct'].agg(['mean','std'])       #여러개를 사용할거면 groupby에 리스트 형식으로 값을 넣으면 가능하다.
```

```python
# 요일별, 시간대별 팁 비율의 횟수, 평균, 표준편차
tips.groupby(['day','time'])['tip_pct'].agg(['count','mean','std'])
```

#### 3. 피벗 테이블(pivot_table)

```python
# 벨류, 인덱스, 컬럼, 함수
# 성별, 흡연유무별 팁의 비율
tips.pivot_table('tip_pct', 'sex', 'smoker', 'mean')
```

```python
# 합계 기능
# 성별, 흡연유무별 인원(count 사용하면 됨)
tips.pivot_table('tip_pct','sex','smoker','count',margins=True, margins_name='계') # margins=True -->'계'를 할거냐 말거냐
```

#### 4. groupby vs pivot_table

- 성별 팁 비율의 평균


```python
tips.groupby('sex')[['tip_pct']].mean() # 시리즈 형식이 싫다면 리스트를 하나 더 붙이면 데이터프레임으로 나온다.
```

```python
tips.pivot_table('tip_pct', 'sex', aggfunc='mean') # 함수를 사용할때는 aggfun=을 사용해야한다.
```

```python
tips.pivot_table('tip_pct', 'sex')      # 평균은 기본으로 나오기때문에 사용하지 않아도 된다.
```

- 흡연 유무에 따른 평균 팁 비율과 최솟값, 최댓값

```python
tips.groupby('smoker')[['tip_pct']].agg(['mean','min','max'])
```

```python
tips.pivot_table('tip_pct','smoker', aggfunc=['mean','min','max'])
```

- 연습 문제 4.7.4

![Alt text](image-6.png)

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# 1. qcut 명령으로 세 개의 나이 그룹을 만든다.
titanic['age_group']= pd.qcut(titanic.age, 3, labels=['청년','중년','노인']).astype(str)
titanic.head()
```

```python
# 2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다.
def survival_rate(a):
    return a.sum() / len(titanic)
titanic.groupby(['sex','class'])['age_group'].agg([survival_rate])
```

```python
# 3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.
titanic.pivot_table('survival_rate','sex','pclass',aggfunc=['mean'])
```
