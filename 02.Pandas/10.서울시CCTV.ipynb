### 2.10 서울시 구별 CCTV 현황 분석
- 서울시 자치구 연도별 CCTV 설치 현황
- 서울시 주민등록인구(구별)통계

```python
import numpy as np
import pandas as pd
```

#### 1.CCTV 데이터

```python
cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_230630기준.csv',
                    encoding='EUC-KR', index_col='구분',thousands=',')              
cctv.head()
```

```python
cctv.drop(columns=['Unnamed: 0','2023년','총계'],inplace=True)
cctv.drop(['계'], inplace=True)    # '계' 행 지우기
cctv.head(3)
```

```python
# 결측치가 있는지 확인
cctv['2014년 이전'].isna().sum()
```

```python
# 결측치를 0으로 대체하고, 타입을 정수로 변경
cctv['2014년 이전'] = cctv['2014년 이전'].fillna(0).astype(int)
cctv.head()
```

```python
cctv['CCTV댓수'] = cctv.sum(axis=1)
cctv.head(3)
```

```python
cctv['최근증가율'] = (cctv.loc[:, '2020년':'2022년'].sum(axis=1) / cctv['CCTV댓수'] * 100).round(2)
cctv.head(3)
```

```python
cctv.drop(columns=['2014년 이전'] + '2014년	2015년	2016년	2017년	2018년	2019년	2020년	2021년	2022년'.split(),
            inplace=True)
cctv.head()
```

```python
#' 중 구' 의 blank 없애기
cctv.index = cctv.index.str.replace(' ', '')
cctv.index[1]
```

#### 2. 인구 데이터 정리하기

```python
ps = pd.read_csv('data/서울시구별주민등록인구_20230912141751.csv',
                skiprows=1)
ps.head()
```

```python
ps.drop(columns=['동별(2)','세대 (세대).1',	'계 (명).1','한국인 (명).1','등록외국인 (명).1','인구밀도'], inplace=True)
ps.head()
```

```python
ps['외국인비율'] = (ps['등록외국인 (명)'] / ps['계 (명)'] * 100).round(2)  
ps.head()
```

```python
ps['고령자비율'] = (ps['65세이상고령자 (명)'] / ps['계 (명)'] * 100).round(2)
ps.head()
```

```python
ps.drop(columns=['세대 (세대)','세대당인구 (명)'], inplace=True)
ps.head()
```

```python
# 센세 풀이
pop = pd.read_csv('data/서울시구별주민등록인구_20230912141751.csv',skiprows=1, thousands=',')
pop.head(5)
```

```python
pop = pop[['동별(2)','계 (명)','한국인 (명)','등록외국인 (명)','65세이상고령자 (명)']]
pop.head(3)
```

```python
pop.columns = ['구별','인구수','내국인','외국인','고령자']
pop.head(3)
```

```python
# 첫번쨰 행 지우기
pop.drop([0], inplace=True)
pop.head(3)
```

```python
# 구별을 인덱스로 만들기
pop.set_index('구별', inplace=True)
pop.head(3)
```

```python
# 외국인 비율
pop['외국인 비율'] = (pop.외국인 / pop['인구수'] * 100).round(2)
pop['고령자 비율'] = (pop.고령자 / pop['인구수'] * 100).round(2)
pop.head(3)
```

#### 3. CCTV 데이터와 POP 데이터 합치기

```python
df = pd.merge(cctv, pop, left_index=True, right_index=True) # 둘다 index이기에 index 사용
df.head()
```

```python
pd.concat([cctv, pop], axis=1)
```

```python
df.index.name = '구별'
df.head()
```

```python
df.to_csv('data/서울시 구별 CCTV 인구 현황.csv')
```
