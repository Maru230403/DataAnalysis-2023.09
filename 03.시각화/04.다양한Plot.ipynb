## 3. 파이 차트

```python
ratio = [10, 15, 20, 28]        # 합이 100이 아니어도 됨
labels = ['감','귤','밤','배']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('파이 차트')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=colors, explode=explode,
        startangle=90, shadow=True)
plt.title('파이 차트')
plt.show()
```

### 5. 산점도(Scatter plot)

```python
import seaborn as sns
iris = sns.load_dataset('iris')
X, Y = iris.sepal_length, iris.sepal_width
plt.scatter(X, Y)
plt.title('산점도 - Iris sepal length vs width')
plt.show()
```

```python
colors = ['red'] * 50 + ['green'] * 50 + ['blue'] * 50
plt.scatter(X, Y, color=colors)
plt.title('산점도 - Iris sepal length vs width')
plt.show()
```

```python
size = 500 * np.random.rand(150)
plt.scatter(X, Y, color=colors, alpha=0.5, s=size)
plt.title('Bubble chart')
plt.show()
```

### 7. Box Plot

```python
# setosa 품종의 각 피처(ssl, ssw, spl, spw)
ssl =iris[iris.species == 'setosa'].sepal_length      # filtering
ssw = iris[iris.species == 'setosa'].sepal_width 
spl = iris[iris.species == 'setosa'].petal_length 
spw = iris[iris.species == 'setosa'].petal_width 
```

```python
plt.boxplot((ssl, ssw, spl, spw),
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.title('Setosa 품종의 Feature')
plt.show()
```
